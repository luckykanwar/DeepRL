```python
import numpy as np
import gym
import random
```

```python
env = gym.make("Taxi-v2")
env.render()
```

```python
action_size = env.action_space.n
state_size = env.observation_space.n
```

```python
print(action_size, state_size)
```

```python
qtable = np.zeros((state_size, action_size))
print(qtable.shape)
```

```python
# Hyperparameters
episodes = 50000
epsilon = 1.0
max_epsilon = 1.0
min_epsilon = 0.01
learning_rate = 0.7
gamma = 0.65
max_steps = 100
decay_rate = 0.01
```

```python
rewards = []

for episode in range(episodes):
    state = env.reset()
    done = False
    total_reward = 0
    step = 0
    
    for step in range(max_steps):
        # What should be my action in the current state
        expl_expr_tradeoff = random.uniform(0,1)

        if expl_expr_tradeoff > epsilon:
            #exploitation of the action with the highest cumulative reward for the current state
            action = np.argmax(qtable[state,:])
        else:
            action = env.action_space.sample()
        
        new_state, reward, done, info = env.step(action)
        #print(reward)
        
        qtable[state,action] = qtable[state, action] + learning_rate * (reward + gamma * np.max(qtable[new_state, :]) - qtable[state, action])
        
        if done == True:
            break;
            
        state = new_state
        total_reward += reward
        
    episode += 1
    epsilon = min_epsilon + (max_epsilon - min_epsilon)*np.exp(-decay_rate*episode)
    rewards.append(total_reward)
    
print(qtable)
print("Total score over time : ", sum(rewards)/episodes)
```

```python
env.reset()
rewards = []

for episode in range(100):
    state = env.reset()
    step = 0
    done = False
    print("****************************************************")
    print("EPISODE ", episode)
    
    total_reward = 0
    for step in range(max_steps):
        #print("step: ", step)
        env.render()
        # Take the action (index) that have the maximum expected future reward given that state
        action = np.argmax(qtable[state,:])
        #print("action: ", action)
        new_state, reward, done, info = env.step(action)
        #print("new_state: ",new_state)
        if done:
            #print("done on step : ", step)
            break
        total_reward += reward
        state = new_state
    print("Score", total_reward)
    rewards.append(total_reward)
env.close()
print("Total score over time : ", sum(rewards)/100)
```
